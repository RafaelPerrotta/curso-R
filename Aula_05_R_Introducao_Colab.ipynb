# đ ExplicaĂ§ĂŁo das Principais MĂŠtricas de Marketing (VisĂŁo de NegĂłcio)

As mĂŠtricas de marketing sĂŁo essenciais para **avaliar o desempenho das campanhas**, tomar decisĂľes estratĂŠgicas e otimizar o retorno sobre o investimento. Abaixo estĂĄ uma explicaĂ§ĂŁo prĂĄtica e orientada a resultados de cada mĂŠtrica.

### 1. Alcance (Reach)
- **O que ĂŠ**: NĂşmero de pessoas Ăşnicas que viram sua publicaĂ§ĂŁo/anĂşncio.
- **ImportĂ˘ncia para o negĂłcio**: Mede o tamanho da audiĂŞncia atingida. Ătil para construir **awareness** (consciĂŞncia de marca).
- **Meta tĂ­pica**: Quanto maior, melhor (depende do objetivo). Bom para campanhas de topo de funil.

### 2. ImpressĂľes
- **O que ĂŠ**: NĂşmero total de vezes que sua publicaĂ§ĂŁo/anĂşncio foi exibido (mesma pessoa pode ver vĂĄrias vezes).
- **ImportĂ˘ncia**: Indica a **frequĂŞncia** de exposiĂ§ĂŁo. Alta impressĂŁo com baixo alcance = pĂşblico saturado.

### 3. CPC â Custo por Clique (Cost Per Click)
- **O que ĂŠ**: Quanto a empresa paga, em mĂŠdia, cada vez que alguĂŠm clica no anĂşncio.
- **VisĂŁo de negĂłcio**: Quanto mais baixo, melhor (maior eficiĂŞncia).
- **Exemplo**: CPC de R$ 2,50 ĂŠ bom no Google Ads; acima de R$ 6,00 pode indicar competiĂ§ĂŁo alta ou criativo ruim.

### 4. CPM â Custo por Mil ImpressĂľes
- **O que ĂŠ**: Custo para exibir o anĂşncio para 1.000 pessoas.
- **ImportĂ˘ncia**: Principal mĂŠtrica para campanhas de **branding** e awareness. Quanto menor, mais eficiente ĂŠ o alcance.

### 5. CAC â Custo de AquisiĂ§ĂŁo de Cliente
- **O que ĂŠ**: Valor total investido em marketing dividido pelo nĂşmero de novos clientes conquistados.
- **Por que ĂŠ crĂ­tica**: Uma das mĂŠtricas mais importantes do negĂłcio.
- **Regra de ouro**: O CAC deve ser **recuperado em menos de 12 meses** (idealmente em 3â6 meses). CAC alto = negĂłcio insustentĂĄvel.

### 6. CTR â Taxa de Cliques (Click-Through Rate)
- **O que ĂŠ**: Porcentagem de pessoas que clicaram no anĂşncio em relaĂ§ĂŁo Ă s impressĂľes.
- **Bom benchmark**:
  - Google Ads: 3â5%
  - Instagram/Facebook: 0,9â2%
  - Email: 2â4%
- **InterpretaĂ§ĂŁo**: Reflete a **relevĂ˘ncia** do anĂşncio.

### 7. Taxa de ConversĂŁo
- **O que ĂŠ**: Porcentagem de pessoas que realizaram a aĂ§ĂŁo desejada (compra, cadastro, download, etc.).
- **ImportĂ˘ncia**: Mede a **eficĂĄcia** do funil. Uma pequena melhora aqui pode gerar grande impacto financeiro.

### 8. ROAS â Return on Ad Spend
- **O que ĂŠ**: Receita gerada Ăˇ Investimento em anĂşncios.
- **Exemplo**: ROAS de 5x = para cada R$ 1 investido, retorna R$ 5 em receita.
- **Meta mĂ­nima recomendada**: 4x (depende da margem do produto).

### 9. Ticket MĂŠdio
- **O que ĂŠ**: Valor mĂŠdio gasto por cliente em uma compra.
- **EstratĂŠgia**: Aumentar o ticket mĂŠdio (upsell, cross-sell) ĂŠ geralmente mais barato que adquirir novos clientes.

### 10. Taxa de Engajamento
- **O que ĂŠ**: InteraĂ§Ăľes (curtidas, comentĂĄrios, compartilhamentos, salvos) Ăˇ Alcance.
- **ImportĂ˘ncia**: Indica o quanto o conteĂşdo **conecta** com o pĂşblico. Alto engajamento melhora o algoritmo das redes.

### 11. Taxa de RejeiĂ§ĂŁo (Bounce Rate)
- **O que ĂŠ**: Percentual de visitantes que entram no site e saem sem interagir.
- **Meta ideal**: Abaixo de 50%. Acima de 70% indica problema de landing page, carregamento lento ou pĂşblico errado.

### 12. Open Rate (Taxa de Abertura de Email)
- **O que ĂŠ**: Percentual de emails abertos.
- **Bom benchmark**: 20â30% (varia por segmento).

| NĂ­vel          | Exemplos de MĂŠtricas                  | ResponsĂĄvel          | Foco Principal          |
|----------------|---------------------------------------|----------------------|-------------------------|
| **PrimĂĄrio**   | CAC, ROAS, Receita                    | C-Level / Diretoria  | Resultado financeiro    |
| **SecundĂĄrio** | CTR, Taxa de ConversĂŁo, CPC           | Gerentes de Marketing| EficiĂŞncia de canais    |

Nunca olhe uma mĂŠtrica isolada. O mais importante ĂŠ entender a **relaĂ§ĂŁo entre elas**.  
Exemplo: Um CAC baixo com ROAS alto = campanha excelente.



```r
# ================================================
# DASHBOARD MĂTRICAS DE MARKETING - VERSĂO 5.3 (FINAL)
# KPI Cards + Funil AIDA + GrĂĄficos + Tabela gt
# ================================================

suppressPackageStartupMessages({
  pacotes <- c("tidyverse", "lubridate", "scales", "patchwork", "ggthemes", "gt")
  for (p in pacotes) {
    if (!require(p, character.only = TRUE, quietly = TRUE)) {
      install.packages(p, quiet = TRUE)
      library(p, character.only = TRUE, quietly = TRUE)
    }
  }
})

# ==================== DADOS ====================
set.seed(123)

datas <- seq(ymd("2025-01-01"), ymd("2025-06-30"), by = "week")

campanhas <- tibble(
  data = rep(datas, each = 4),
  canal = rep(c("Instagram", "Google Ads", "Email", "Facebook"), length(datas)),
  orcamento = round(rnorm(length(datas)*4, 8500, 2500), 0),
  impressoes = round(rnorm(length(datas)*4, 85000, 35000), 0),
  alcance = round(impressoes * runif(length(datas)*4, 0.65, 0.92), 0),
  cliques = round(impressoes * runif(length(datas)*4, 0.012, 0.085), 0),
  conversoes = round(cliques * runif(length(datas)*4, 0.03, 0.22), 0),
  custo = round(orcamento * runif(length(datas)*4, 0.85, 1.15), 0),
  ticket_medio = round(rnorm(length(datas)*4, 185, 45), 0),
  engajamento = round(impressoes * runif(length(datas)*4, 0.008, 0.045), 0)
) %>%
  mutate(
    CPC = round(custo / cliques, 2),
    CPM = round((custo / impressoes) * 1000, 2),
    CAC = round(custo / conversoes, 2),
    CTR = round((cliques / impressoes) * 100, 2),
    Taxa_Conversao = round((conversoes / cliques) * 100, 2),
    Bounce_Rate = round(runif(n(), 35, 68), 1),
    Open_Rate = if_else(canal == "Email", round(runif(n(), 18, 42), 1), NA_real_),
    ROAS = round((conversoes * ticket_medio) / custo, 2),
    Engajamento_Rate = round((engajamento / alcance) * 100, 2)
  )

# ==================== KPI CARDS ====================
kpi_geral <- campanhas %>%
  summarise(
    Total_Investido = sum(custo),
    CAC_Geral = round(mean(CAC), 2),
    ROAS_Geral = round(mean(ROAS), 2),
    Conversao_Media = round(mean(Taxa_Conversao), 2)
  )

create_card <- function(titulo, valor, subtitulo = "", cor = "#2C3E50") {
  ggplot() +
    annotate("rect", xmin = 0, xmax = 1, ymin = 0, ymax = 1,
             fill = cor, alpha = 0.08, color = cor, size = 1.2) +
    annotate("text", x = 0.5, y = 0.68, label = titulo, size = 5, fontface = "bold", color = cor) +
    annotate("text", x = 0.5, y = 0.38, label = valor, size = 9.5, fontface = "bold") +
    annotate("text", x = 0.5, y = 0.15, label = subtitulo, size = 4, color = "gray40") +
    theme_void() +
    theme(plot.margin = margin(10, 10, 10, 10))
}

card1 <- create_card("Investimento Total",
                     scales::dollar(kpi_geral$Total_Investido, big.mark = ".", decimal.mark = ","),
                     "JanâJun 2025", "#34495E")
card2 <- create_card("CAC MĂŠdio",
                     scales::dollar(kpi_geral$CAC_Geral, big.mark = ".", decimal.mark = ","),
                     "Custo por Cliente", "#E74C3C")
card3 <- create_card("ROAS", paste0(kpi_geral$ROAS_Geral, "x"),
                     "Retorno sobre Investimento", "#27AE60")
card4 <- create_card("Taxa de ConversĂŁo", paste0(kpi_geral$Conversao_Media, "%"),
                     "MĂŠdia Geral", "#3498DB")

# ==================== FUNIL AIDA ====================
funil_aida <- tibble(
  Etapa = c("AtraĂ§ĂŁo (Awareness)", "Interesse (Interest)",
            "Desejo (Desire)", "AĂ§ĂŁo (Action)"),
  Quantidade = c(sum(campanhas$alcance), sum(campanhas$cliques),
                 round(sum(campanhas$cliques) * 0.45), sum(campanhas$conversoes)),
  Percentual = c(100, 28, 12, 4.5)
) %>%
  mutate(Ordem = row_number())

p_aida <- funil_aida %>%
  ggplot(aes(x = reorder(Etapa, -Ordem), y = Quantidade)) +
  geom_col(aes(fill = Etapa), width = 0.65) +
  geom_text(aes(label = paste0(scales::comma(Quantidade), "\n(", Percentual, "%)")),
            color = "white", size = 5, fontface = "bold", vjust = 0.5) +
  scale_fill_manual(values = c("#1ABC9C", "#3498DB", "#9B59B6", "#E74C3C")) +
  coord_flip() +
  labs(title = "FUNIL AIDA", subtitle = "Jornada do Cliente", x = NULL, y = "Volume") +
  theme(axis.text.y = element_text(size = 13, face = "bold"),
        legend.position = "none")

# ==================== GRĂFICOS (ajustados para melhor legibilidade) ====================
p1 <- campanhas %>%
  group_by(data, canal) %>%
  summarise(Alcance = sum(alcance), Impressoes = sum(impressoes), .groups = "drop") %>%
  pivot_longer(c(Alcance, Impressoes)) %>%
  ggplot(aes(data, value/1000, color = name)) +
  geom_line(size = 1.3) + geom_point(size = 2.5) +
  facet_wrap(~canal, scales = "free_y", ncol = 2) +
  scale_y_continuous(labels = label_comma(suffix = "K")) +
  labs(title = "Alcance Ă ImpressĂľes", x = NULL, y = "Quantidade (milhares)", color = NULL) +
  theme(legend.position = "top")

p2 <- campanhas %>%
  group_by(canal) %>%
  summarise(CPC = mean(CPC), CPM = mean(CPM), CAC = mean(CAC)) %>%
  pivot_longer(c(CPC, CPM, CAC)) %>%
  ggplot(aes(canal, value, fill = name)) +
  geom_col(position = "dodge", width = 0.7) +
  geom_text(aes(label = paste0("R$ ", round(value,1))),
            position = position_dodge(0.7), vjust = -0.5, size = 4.5, fontface = "bold") +
  scale_fill_brewer(palette = "Set2") +
  labs(title = "CPC, CPM e CAC por Canal", y = "Valor (R$)") +
  theme(axis.text.x = element_text(angle = 30, hjust = 1))

p3 <- campanhas %>%
  filter(canal == "Email" | !is.na(Open_Rate)) %>%
  group_by(canal) %>%
  summarise(CTR = mean(CTR), Conversao = mean(Taxa_Conversao), Open = mean(Open_Rate, na.rm = TRUE)) %>%
  pivot_longer(c(CTR, Conversao, Open), names_to = "Taxa", values_to = "Percentual") %>%
  ggplot(aes(canal, Percentual, fill = Taxa)) +
  geom_col(position = "dodge", width = 0.75) +
  geom_text(aes(label = paste0(round(Percentual,1), "%")),
            position = position_dodge(0.75), vjust = -0.5, size = 4.5, fontface = "bold") +
  scale_fill_brewer(palette = "Set1") +
  labs(title = "Taxas de Performance (%)", y = NULL)

p4 <- campanhas %>%
  group_by(canal) %>%
  summarise(Engajamento = mean(Engajamento_Rate), Rejeicao = mean(Bounce_Rate)) %>%
  ggplot(aes(Engajamento, Rejeicao, color = canal)) +
  geom_point(size = 6) +
  geom_text(aes(label = canal), vjust = -1.8, size = 4.8, fontface = "bold") +
  scale_x_continuous(labels = percent_format(scale = 1)) +
  scale_y_continuous(labels = percent_format(scale = 1)) +
  labs(title = "Engajamento vs RejeiĂ§ĂŁo", x = "Engajamento (%)", y = "Bounce Rate (%)") +
  theme(legend.position = "none")

p5 <- campanhas %>%
  group_by(canal) %>%
  summarise(ROAS = mean(ROAS), Ticket = mean(ticket_medio)) %>%
  ggplot(aes(Ticket, ROAS, color = canal)) +
  geom_point(size = 6) +
  geom_text(aes(label = canal), vjust = -1.8, size = 4.8, fontface = "bold") +
  scale_x_continuous(labels = dollar) +
  labs(title = "ROAS Ă Ticket MĂŠdio", x = "Ticket MĂŠdio (R$)", y = "ROAS")

# ==================== TABELA GT ====================
tabela_resumo <- campanhas %>%
  group_by(canal) %>%
  summarise(
    `CPC (R$)` = round(mean(CPC), 2),
    `CPM (R$)` = round(mean(CPM), 2),
    `CAC (R$)` = round(mean(CAC), 2),
    `CTR (%)` = round(mean(CTR), 2),
    `Taxa ConversĂŁo (%)` = round(mean(Taxa_Conversao), 2),
    `Engajamento (%)` = round(mean(Engajamento_Rate), 2),
    `ROAS` = round(mean(ROAS), 2)
  ) %>%
  gt(rowname_col = "canal") %>%
  tab_header(
    title = "RESUMO GERAL POR CANAL",
    subtitle = "MĂŠdias do perĂ­odo JanâJun 2025"
  ) %>%
  tab_spanner(label = "Custos", columns = c(`CPC (R$)`, `CPM (R$)`, `CAC (R$)`)) %>%
  tab_spanner(label = "Performance", columns = c(`CTR (%)`, `Taxa ConversĂŁo (%)`)) %>%
  fmt_currency(columns = c(`CPC (R$)`, `CPM (R$)`, `CAC (R$)`), currency = "BRL", decimals = 2) %>%
  fmt_number(columns = c(`CTR (%)`, `Taxa ConversĂŁo (%)`, `Engajamento (%)`, ROAS), decimals = 2) %>%
  tab_options(table.font.size = px(14), heading.title.font.size = px(16))

# ==================== DASHBOARD FINAL ====================
cards <- (card1 + card2 + card3 + card4) + plot_layout(ncol = 4)

dashboard <- cards / (p1 + p_aida) / (p2 + p3) / (p4 + p5) / tabela_resumo +
  plot_annotation(
    title = "DASHBOARD COMPLETO DE MĂTRICAS DE MARKETING",
    subtitle = "KPI Cards + Funil AIDA + GrĂĄficos + Tabela",
    caption = "Projeto em R â˘ Grok 2026 â˘ Baseado no conteĂşdo ESPM",
    theme = theme(plot.title = element_text(size = 22, face = "bold", hjust = 0.5))
  )

print(dashboard)
print(tabela_resumo)

# ==================== EXPORTAĂĂO ====================
ggsave("dashboard_metricas_marketing_v5.3.png",
       dashboard, width = 19, height = 24, dpi = 400, bg = "white")

# Salvar tabela como imagem (com fallback)
tryCatch({
  gtsave(tabela_resumo, "tabela_resumo_metricas.png")
  cat("Tabela salva como imagem!\n")
}, error = function(e) {
  cat("Nota: webshot2 nĂŁo instalado. Tabela salva apenas como HTML.\n")
  gtsave(tabela_resumo, "tabela_resumo_metricas.html")
})

cat("â Dashboard v5.3 gerado com sucesso!\n")
cat("Arquivo principal: dashboard_metricas_marketing_v5.3.png\n")
```
