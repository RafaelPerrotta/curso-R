## 1. Operações Matemáticas Básicas
O R pode ser usado como uma calculadora poderosa.

```r
# Exemplo 1: Adição e Subtração
soma <- 150 + 235
subtracao <- 500 - 120
print(paste("Soma:", soma))
print(paste("Subtração:", subtracao))
```

```r
# Exemplo 2: Multiplicação, Divisão e Potenciação
multiplicacao <- 25 * 4
divisao <- 100 / 5
potencia <- 2^8
print(paste("Multiplicação:", multiplicacao))
print(paste("Divisão:", divisao))
print(paste("Potência:", potencia))
```

```r
# Exemplo 3: Funções matemáticas integradas
raiz_quadrada <- sqrt(144)
logaritmo <- log(100) # logaritmo natural (base e)
log_base10 <- log10(100)
print(paste("Raiz Quadrada:", raiz_quadrada))
print(paste("Logaritmo (base e):", logaritmo))
print(paste("Logaritmo (base 10):", log_base10))
```

## 2. Variáveis e Tipos de Dados
Atribuição de valores e verificação de tipos.

```r
# Exemplo 4: Atribuição de Variáveis
nome <- "Ciência de Dados"
ano <- 2024
ativo <- TRUE
print(nome)
print(ano)
print(ativo)
```

## 3. Estruturas de Dados: Vetores
Vetores são coleções de elementos do mesmo tipo.

```r
# Exemplo 6: Criando vetores com a função c()
idades <- c(25, 30, 22, 28, 35)
nomes_alunos <- c("Ana", "Bruno", "Carlos", "Diana", "Eduardo")
print(idades)
print(nomes_alunos)
```

```r
# Exemplo 7: Operações com vetores (Vetorização)
idades_em_5_anos <- idades + 5
print("Idades atuais:")
print(idades)
print("Idades daqui a 5 anos:")
print(idades_em_5_anos)
```

```r
# Exemplo 8: Acessando elementos de um vetor (Indexação começa em 1 no R)
primeiro_aluno <- nomes_alunos[1]
tres_primeiros <- nomes_alunos[1:3]
print(primeiro_aluno)
print(tres_primeiros)
```

## 4. Estruturas de Dados: Matrizes e Listas

```r
# Exemplo 9: Criando uma Matriz
minha_matriz <- matrix(1:9, nrow = 3, ncol = 3, byrow = TRUE)
print("Matriz 3x3:")
print(minha_matriz)
```

```r
# Exemplo 10: Criando uma Lista (pode conter diferentes tipos de dados)
perfil_aluno <- list(
  nome = "Maria",
  idade = 27,
  notas = c(8.5, 9.0, 7.5),
  aprovado = TRUE
)
print(perfil_aluno)
print(paste("Acessando as notas da lista:", paste(perfil_aluno$notas, collapse=", ")))
```

## 5. Data Frames
A estrutura mais importante para Ciência de Dados em R.

```r
# Exemplo 11: Criando um Data Frame
dados_turma <- data.frame(
  ID = 1:5,
  Nome = c("Ana", "Bruno", "Carlos", "Diana", "Eduardo"),
  Idade = c(25, 30, 22, 28, 35),
  Nota_Final = c(9.2, 7.5, 8.8, 9.5, 6.0)
)
print(dados_turma)
```

```r
# Exemplo 12: Explorando o Data Frame
str(dados_turma)      # Estrutura do data frame
summary(dados_turma)  # Resumo estatístico das colunas
```

```r
# Exemplo 13: Acessando colunas de um Data Frame
notas <- dados_turma$Nota_Final
print("Notas da turma:")
print(notas)
```

---

## 6. Estatística Descritiva Básica

```r
# Exemplo 14: Funções estatísticas comuns
media_notas <- mean(dados_turma$Nota_Final)
mediana_notas <- median(dados_turma$Nota_Final)
desvio_padrao <- sd(dados_turma$Nota_Final)
nota_maxima <- max(dados_turma$Nota_Final)
nota_minima <- min(dados_turma$Nota_Final)

print(paste("Média:", media_notas))
print(paste("Mediana:", mediana_notas))
print(paste("Desvio Padrão:", round(desvio_padrao, 2)))
print(paste("Nota Máxima:", nota_maxima))
print(paste("Nota Mínima:", nota_minima))
```

## 7. Operadores Lógicos e Filtragem de Dados

```r
# Exemplo 15: Operadores Lógicos
print("Alunos com nota maior que 8:")
dados_turma$Nota_Final > 8
```

```r
# Exemplo 16: Filtrando um Data Frame
alunos_aprovados <- subset(dados_turma, Nota_Final >= 7.0)
print("Alunos Aprovados (Nota >= 7.0):")
print(alunos_aprovados)
```

## 8. Estruturas de Controle

```r
# Exemplo 17: Condicional IF / ELSE
nota_exemplo <- 8.5
if(nota_exemplo >= 7) {
  print("Status: Aprovado")
} else {
  print("Status: Reprovado")
}
```

```r
# Exemplo 18: Loop FOR
print("Quadrado dos números de 1 a 5:")
for(i in 1:5) {
  print(paste(i, "^ 2 =", i^2))
}
```

## 9. Criando Funções Personalizadas

```r
# Exemplo 19: Definindo uma função em R
calcular_imc <- function(peso, altura) {
  imc <- peso / (altura^2)
  return(round(imc, 2))
}

meu_imc <- calcular_imc(peso = 75, altura = 1.75)
print(paste("O IMC calculado é:", meu_imc))
```

## 10. Visualização Básica de Dados
Usando as funções gráficas nativas do R e um dataset embutido.

```r
# Exemplo 20: Carregando um dataset nativo (mtcars)
data(mtcars)
head(mtcars, 3) # Mostra as 3 primeiras linhas
```

```r
# Exemplo 21: Gráfico de Dispersão (Scatter Plot)
plot(mtcars$wt, mtcars$mpg,
     main = "Relação entre Peso e Consumo de Combustível",
     xlab = "Peso (1000 lbs)",
     ylab = "Milhas por Galão (mpg)",
     col = "blue",
     pch = 19) # pch=19 usa círculos sólidos
```

```r
# Exemplo 22: Histograma
hist(mtcars$mpg,
     main = "Distribuição do Consumo de Combustível",
     xlab = "Milhas por Galão (mpg)",
     ylab = "Frequência",
     col = "lightgreen",
     breaks = 10)
```

```r
# Exemplo 23: Boxplot
boxplot(mpg ~ cyl, data = mtcars,
        main = "Consumo de Combustível por Número de Cilindros",
        xlab = "Número de Cilindros",
        ylab = "Milhas por Galão (mpg)",
        col = c("salmon", "lightblue", "orange"))
```

## 11. Introdução a Pacotes
O verdadeiro poder do R está nos seus pacotes. (Apenas ilustrativo, pois no Colab alguns já vêm instalados).

```r
# Exemplo 24: Instalando e carregando um pacote (ggplot2)
# install.packages("ggplot2") # Descomente para instalar se necessário
library(ggplot2)

# Criando um gráfico mais avançado com ggplot2
ggplot(mtcars, aes(x = wt, y = mpg, color = as.factor(cyl))) +
  geom_point(size = 3) +
  labs(title = "Gráfico Avançado: Peso vs Consumo",
       x = "Peso",
       y = "Consumo (mpg)",
       color = "Cilindros") +
  theme_minimal()
```
