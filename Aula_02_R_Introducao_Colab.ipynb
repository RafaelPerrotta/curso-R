# Aula 02 - Distribuições Estatísticas em R

Nesta aula, exploramos distribuições importantes:
- Normal
- Poisson
- Exponencial
- Weibull
- Gamma


## 1. Distribuição Normal
Simula alturas de pessoas em uma população.

```r
set.seed(123)
dados_normal <- rnorm(1000, mean=170, sd=10)
hist(dados_normal, main='Distribuição Normal', xlab='Altura', col='lightblue', border='black')
```

## 2. Distribuição de Poisson
Número de chamadas recebidas por minuto em um call center.

```r
dados_poisson <- rpois(1000, lambda=5)
hist(dados_poisson, main='Distribuição de Poisson', xlab='Chamadas', col='lightgreen', border='black')
```

## 3. Distribuição Exponencial
Tempo de espera entre eventos (ex: chegada de clientes).

```r
dados_exp <- rexp(1000, rate=0.2)
hist(dados_exp, main='Distribuição Exponencial', xlab='Tempo', col='lightcoral', border='black')
```

## 4. Distribuição Weibull
Usada para modelar tempo de vida de equipamentos.

```r
dados_weibull <- rweibull(1000, shape=2, scale=5)
hist(dados_weibull, main='Distribuição Weibull', xlab='Tempo de vida', col='lightyellow', border='black')
```

## 5. Distribuição Gamma
Modela tempo até ocorrência de múltiplos eventos.

```r
dados_gamma <- rgamma(1000, shape=2, rate=0.5)
hist(dados_gamma, main='Distribuição Gamma', xlab='Tempo acumulado', col='lightpink', border='black')
```
